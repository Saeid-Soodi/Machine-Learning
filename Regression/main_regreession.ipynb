```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data
# y = 2x-1
x= np.array([1,2,3,4]).reshape(-1,1)
y=np.array([1,3,5,7])
```

```python
#plot 
plt.plot(x,y,'x')
plt.show()
```

```python
#model
from sklearn.linear_model import LinearRegression
model =LinearRegression()
model.fit(x,y)
```

```python
#test  y =m x * x +c
m = model.coef_[0]
c=model.intercept_
print(m,c)
```

```python
#plot 
plt.plot(x,y,'x')
plt.plot(x,model.predict(x))
plt.show()
```

```python
#result of model for data set 
result = model.predict([[5]])
result
```

```python
# Example  relation between year of work and salires for starting in regression
year_of_experience = np.array([0,1,2,3,4,6,8,10]).reshape(-1,1)
salaries=np.array([30000,40000,55000,60000,70000,80000,85000,87000])
plt.plot(year_of_experience,salaries, 'x')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
model =DecisionTreeRegressor(max_depth=4)
model.fit(year_of_experience,salaries)
y_predict=model.predict(year_of_experience)
plt.plot(year_of_experience,salaries,'*')
plt.plot(year_of_experience,y_predict)
plt.show()

```
