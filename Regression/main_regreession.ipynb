```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data
# y = 2x-1
x= np.array([1,2,3,4]).reshape(-1,1)
y=np.array([1,3,5,7])
```

```python
#plot 
plt.plot(x,y,'x')
plt.show()
```

```python
#model
from sklearn.linear_model import LinearRegression
model =LinearRegression()
model.fit(x,y)
```

```python
#test  y =m x * x +c
m = model.coef_[0]
c=model.intercept_
print(m,c)
```

```python
#plot 
plt.plot(x,y,'x')
plt.plot(x,model.predict(x))
plt.show()
```

```python
#result of model for data set 
result = model.predict([[5]])
result
```
