# working with MatplotLib library

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[3,45,56,7,8,9,0,7,1]
plt.plot(x,y,color="red")
plt.title("Test")
plt.xlabel("time")
plt.ylabel("money")
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[3,45,56,7,8,9,0,7,1]
plt.plot(x,y,"x",color="red")
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[3,45,56,7,8,9,0,7,1]
plt.plot(x,y,marker="x",color="red")
```

```python
#bar 
x=["green","red"]
y=[2,10]
plt.bar(x,y)
plt.show()
```

```python
#pie chart
x=["green","red","yellow","blue"]
y=[2,10,15,4]
plt.pie(y,labels=x, autopct="%1.1f%%",colors=["green","red","yellow","blue"])
plt.show()
```

```python
import numpy as np
```

```python
x=np.random.rand(1000,1)
plt.hist(x)
plt.show()
```
